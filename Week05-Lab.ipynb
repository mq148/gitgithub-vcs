### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
# Thersold Answer 3:

with open("temps.txt", "w") as file: # opening the file with write mode and with closes the file.
    for day in range(1, 6): # runs a loop to to represent 5 days.
        temp = float(input(f"Enter temperature for the Day please {day}: "))
        file.write(f"Day {day}: {temp}°C\n") # writing each day's tempreature in the file.


with open("temps.txt", "r") as file: # read mode open.
    lines = file.readlines() # stores the list of lines in the file.

total_temp = 0
for line in lines:
    temp_str = line.strip().split(": ")[1].replace("°C", "") # "line.strip removes the whitespaces, split() splits line in two parts, , replace give numeric."
    total_temp += float(temp_str) # incrementing the total and storing in total_temp.

average_temp = total_temp / len(lines) # calculating the avg.


print("\nRecorded Temperatures are:")
for line in lines:
    print(line.strip()) # Display all recorded temperatures without extra spaces.

print(f"\nAverage Temperature: {average_temp:.2f}°C")
`````

### **Activity 5: Unique Hobbies Set** 


Two friends share their hobbies:
alex_hobbies = {"reading", "gaming", "cooking", "hiking"}
sam_hobbies = {"gaming", "hiking", "photography", "music"}

Find and print:
- Hobbies both enjoy (intersection)
- Hobbies only Alex enjoys
- All unique hobbies combined

---

`````python
# Thersold Answer 5:

# Defining the hobby sets as per details of the question.
alex_hobbies = {"reading", "gaming", "cooking", "hiking"}
sam_hobbies = {"gaming", "hiking", "photography", "music"}

both_enjoy = alex_hobbies & sam_hobbies # using set intersection opertator (&) to find comm
on hobbies.

alex_hobby = alex_hobbies - sam_hobbies # finding alex favourite hobby by subtracting and storing in varibale.

all_unique_hobbies = alex_hobbies | sam_hobbies # Combining the two sets with there values together.

print("Hobbies both enjoy are:", both_enjoy)
print("Hobbies only Alex enjoys is:", alex_hobby)
print("All unique hobbies combined for Alex & Sam are:", all_unique_hobbies)
`````


### **Activity 6: Simple Grade Book Dictionary** 

Create a dictionary for 3 students and their scores:
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

Perform these tasks:
- Add a new student "Diana" with score 88
- Update Bob's score to 95
- Print all students with scores above 80
- Calculate and print the average score

---

`````python
# Thersold Answer 6:

# Creating a dictionary of students and storing socres as per question. 
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

grades["Diana"] = 88 # Adding the new student given by question.


grades["Bob"] = 95 # Refrencing the dictionary and updating bob's score.

# Displaying students scores > 80:
print("The students who scored above 80 Are:")
for name, score in grades.items():
    if score > 80: # selecting and displaying the students who scored above 80.
        print(f"- {name}: {score}")
        

average_score = sum(grades.values()) / len(grades) # Calculating avergae of all student by dividing by score.
print(f"\nAverage score of students are: {average_score:.2f}")
`````

### **Activity 7: Contact Book** 


Create a contact dictionary:
contacts = {
    "Mom": "555-1234",
    "Dad": "555-5678",
    "Friend": "555-9012"
}

Write a program that:
- Adds a new contact
- Looks up a contact by name
- Prints all contacts in format "Name: Number"

---

`````python
# Thersold Answer 7:

contacts = { # creating a dictionary and storing the key value pairs provided by the question.
    "Mom": "555-1234",
    "Dad": "555-5678",
    "Friend": "555-9012"
}

new_name = input("Enter new contact name please: ")
new_number = input("Enter the phone number please: ")
contacts[new_name] = new_number # adding the new key vlaue pair in the dictionary.


lookup_name = input("\nEnter a name to look up: ")
if lookup_name in contacts:  # The code to check the users key name is found in the contacts dictionary.
    print(f"{lookup_name}'s number is {contacts[lookup_name]}")
else:
    print("Contact is not found in the contacts dictionary.")

print("\nAll contacts Displayed:")
for name, number in contacts.items(): # Runs loop through every key-value pair in the contacts dictionary and return it.
    print(f"{name}: {number}")
`````
