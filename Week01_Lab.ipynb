Typical 1. Total Time to Watch a Series

Compute total viewing time (no breaks) for a series:
- 5 seasons, 10 episodes per season
- Season 1: 45 minutes 35 seconds per episode
- Seasons 2–5: 35 minutes 15 seconds per episode
Display the total as hours, minutes, and seconds.

**Example Output:**
```
Total time: 39h 30m 0s
```


````python
# Typical Answer 1:

# Season 1 calculation:
episodes_season1 = 10
restseason_episodes = 4*10

season1_1episode_timings = (45 * 60) + 35   # Converting min to sec to avoid getting answer in fractions.
season1_totaltime = episodes_season1 * season1_1episode_timings # Finding total time in sec for season 1.

# Rest 4 seasons calculation:
fourseasons_episodes= 4 * 10   # Calculating total episodes.
totaltime_fourseasons_1episode= (35 * 60) + 15   # Finding the total time for one episode in the rest 4 seasons.
totaltime_fourseasons= fourseasons_episodes * totaltime_fourseasons_1episode

total_seconds =season1_totaltime + totaltime_fourseasons # Calculating the total sec of all seasons (111950).

# Finding hours, minutes & seconds:
hours = total_seconds // 3600 # "using // to get ans as a whole number and / totalseconds by 3600 to find hrs."
minutes = (total_seconds % 3600) // 60 # finding remainder of sec and / by to get minutes.
seconds = total_seconds % 60 # finding the leftover seconds after removing full minutes

print(f"Total time: {hours}h {minutes}m {seconds}s") # output in required format.
````

### Excellent 2. Data Science: Mini Statistics Report

Ask for five numeric values (space-separated). Print the values, total, mean, median, and sample standard deviation.

**Example Output:**
```
Values: 80, 90, 70, 85, 75
Total = 400.00
Mean = 80.00
Median = 80.00
Std Dev = 7.91
```


*Optional hint:* `import statistics as stats`


````python
# Excellent Answer 2:

import statistics

input_values = input("Please input 5 numbers ensuring they have spacing: ") # taking input from the user in spaces.
numbers = input_values.split()  # "using split function to seperate the strings into different parts in a list."

values = [] # declaring an empty list to store inputs here.
for num in numbers: # using a for loop to go from the first to the last number.
    values.append(float(num)) # "append is to store the value at the end of the list, float converts our num into decimal and val is list name"

total = sum(values) # using sum to add all values.
mean = statistics.mean(values) 
median = statistics.median(values) # using statistics module.
std_dev = statistics.stdev(values)
 
print(" The Values you entered are:", values) 
print(f"Total = {total:.2f}")
print(f"Mean = {mean:.2f}") # prinitng the outputs for total, mean, median, standard deviation to two decimal places.
print(f"Median = {median:.2f}")
print(f"Std Dev = {std_dev:.2f}")
````

Ask for a file or directory path and print its absolute path, parent, name, stem, and suffix.

````python
# Excellent Answer 3:

from pathlib import Path # importing path class from module pathlib.

directory_path = input("Dear user please enter a directory path: ")

path = Path(directory_path) # using path module with the input(directory path) from user so path function can be applied.

# () paranthese to call a function.
print("The Absolute path is:", path.resolve()) # resolve converts the path to an absolute path.
print("The Parent is:", path.parent) # Prints the parent directory of the path.
print("The Name is:", path.name) # Prints the name of the file.
print("The Stem is:", path.stem) # path.stem will give the file name.
print("The Suffix is:", path.suffix) # Prints the file extension.
````
