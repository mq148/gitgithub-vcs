3. Create a function taking one parameters for a password. The function would convert each character of the password into corresponding ASCII numbers and return one string composed of the ASCII numbers instead

````python
# Thersold Answer 3:

# Function to convert password to ASCII numbers
def password_to_ascii(password):
    ascii_string = ""
    for char in password:
        ascii_string += str(ord(char))  # Converting each character to ASCII and append
    return ascii_string

# Ask user for password
user_password = input("Enter your password please: ")

# Convert password to ASCII string
ascii_password = password_to_ascii(user_password)

print(f"Your password in ASCII numbers: {ascii_password}")
````

4. Create a function (or set of functions ) taking one parameter for a password.

The function (s) will print True if

  - the password contains at least one upper case letter,   
  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‘@’.
- Extra (check if the password contains space. If so, print a suitable error message)

````python
# Thersold Answer 4:

def validate_password(password): 
    # Check for uppercase, lowercase, digit, special character, and spaces
    has_upper = False
    has_lower = False # Defining variables to keep track of the rules for checking the password.
    has_digit = False
    has_special = False
    special_characters = "@!#$%^&*()_+=-[]{}|;:,.<>?/~`" # assigning common characters in variable.

    for char in password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True   # This loop scans each character in the password and cecks for the rules to meet for the password.
        elif char.isdigit():
            has_digit = True
        elif char in special_characters:
            has_special = True
        elif char.isspace():
            print("Password should not contain spaces please.")
            return False  # Stops checking if the space is found.

    if has_upper and has_lower and has_digit and has_special: # runs an if statment to check the conditions and results.
        return True  
    else:
        return False

user_password = input("Enter your password please: ")

# if statment to display the user the results based on the input.
if validate_password(user_password):
    print("Your Password is valid")
else:
    print("Sorry your Password is invalid")
````

12. Password Validation with Multiple Criteria: Ask the user for a password and validate whether it:
- Has more than 8 characters.
- Contains at least one uppercase letter.
- Contains at least one lowercase letter.
- Contains at least one number.
- Contains at least one special character (@, #, $, etc.).
- Extra: If the password contains a space, print an error message

````python
# Typical ANswer 3:

def validate_password(password):
    # Checking for spaces first
    if " " in password:
        print("Error: Password should not contain spaces please!")
        return

    # conditional statment for checking the length.
    if len(password) <= 8:
        print("Password must be more than 8 characters long.")
        return

    has_upper = False  # initialising the flags.
    has_lower = False
    has_digit = False
    has_special = False
    

    for char in password: # running a for loop to check the password the conditions.
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_digit = True
        elif not char.isalnum():  # check for a special character
            has_special = True

    # Check all conditions
    if has_upper and has_lower and has_digit and has_special: # conditional statment ensures that the password meets all the rules:
        print("Password is valid!")
    else:
        print("Password is invalid. Please include:")
        if not has_upper:
            print("- At least one uppercase letter")
        if not has_lower:
            print("- At least one lowercase letter")
        if not has_digit:
            print("- At least one number")
        if not has_special:
            print("- At least one special character (@, #, $, etc.)")


user_password = input("Enter your password: ")
validate_password(user_password)
````

Randomly appends a special character (e.g., @, #, $, etc.) at the end of the secret word.

```python
special_chars = "@#$%^&*!"
```

14. Advanced Student Number Validation: Validate the student number to ensure:
- It starts with "M00".
- It contains exactly 9 characters.
- It has no special characters or spaces

````python
# Typical Answer 5:

def validate_student_number(student_number):
    # Checking if it starts with "M00"
    if not student_number.startswith("M00"):
        print("Student number must start with 'M00'.")
        return

    # Checking the length (exactly 9 characters)
    if len(student_number) != 9:
        print("Student number should be exactly 9 characters long.")
        return

    # Conditional statment for special characters or spaces
    if not student_number.isalnum(): # checks if all characters in the string are alphanumeric.
        print("Student number must not contain special characters or spaces.")
        return

    # If all conditions are true so outputing:
    print("Student number is valid!")


student_number = input("Enter your student number please: ")
validate_student_number(student_number)
````
