4. **Body mass index (BMI)** is a measure of health and weight.

It can be calculated by taking your weight in kilograms and dividing it by the square of your height in meters. The interpretation of BMI for people 20 years or older is as follows:


BMI	Interpretation
BMI < 18.5	Underweight
18.5 ≤ BMI ≤ 25.0	Normal
25.0 ≤ BMI ≤ 30.0	Overweight
30.0 ≤ BMI 	Obese
 
Write a program that prompts the user to enter weight in pounds and height in inches and display the BMI. Note that one pound is 0.45359237 kilograms, and one inch is 0.0254 meters


````python
# Thersold Answer 4:

weight_pounds = float(input("Enter your weight in pounds please: "))
height_inches = float(input("Enter your height in inches please: ")) # taking input from users in float data type fo calculation.

weightkg = weight_pounds * 0.45359237 # "formula and multiplication of pound, inches in storing weight/height kg/m"
heightm = height_inches * 0.0254


bmi = weightkg / (heightm ** 2) # Making the formula for calculating the bmi. for accuracy

# if-elif conditional statment:

if bmi < 18.5: 
    category = ("Underweight")
elif 18.5 <= bmi <= 25.0: # categoring all weight types using there provided values.
    category = ("Normal")
elif 25.0 < bmi <= 30.0:
    category = ("Overweight")
else:
    category = ("Obese")


print(f"Your BMI is: {bmi:.2f} and based on your bmi your category is: {category}.") # Printing bmi to 2 dp with the assigned category.

````

5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
# Thersold Answer 5:

marks = [75, 45, 40, 80, 85, 90, 100, 200, 35, 150] # storing random values in a list called marks.

total = 0 # initialzing total = 0

for mark in marks: # using a for loop for marks.
    
    total = total + mark # calculating total mark as sum of list.
average = total / len(marks) # "calculating avg as total / marks and using length fucntion to find the number of items"

highest = max(marks) 
lowest = min(marks) # getting the highest and lowest values in marks.

passed = 0
failed = 0 # initialzing pass & fail to 0 to count how many pass/fail.

for mark in marks: # using for loop.
    
    if mark >= 50:
        passed = passed + 1  # incrementing the passes count by 1.
    else:
        failed = failed + 1 # increasing the fail count by 1.

print(f"The Average mark is: {average:.2f}") # printing the result in 2 dp.
print(f"The Highest mark is: {highest}")
print(f"The Lowest mark is: {lowest}")
print(f"The Number of students passed: {passed}")
print(f"The Number of students failed: {failed}")
````

17. **Collecting Positive Numbers**
You are writing a small program to help a teacher keep track of positive numbers entered by students. The teacher also wants to know if any even numbers were given.

**Requirements:**

1. Keep asking the user to enter a positive number, or type `"done"` to stop.
2. Store all entered numbers in a **list**.
3. After the input ends:

   * If numbers were entered:

     * Print all the numbers entered.
     * Check the list and print the **first even number**. If no even number exists, print `"No even numbers found"`.
   * If no numbers were entered, print `"No numbers were entered"`.


> ### Note 
>
>* **`lower()`**: Converts any text into lowercase.
>  Example: `"DONE".lower()` becomes `"done"`.
>  This makes it easier to compare user input without worrying about uppercase/lowercase.
>
>* **`isdigit()`**: Checks if a string only contains digits (0–9).
>  Example: `"123".isdigit()` returns `True`, but `"12a".isdigit()` returns `False`.
>  This helps us make sure the user typed a valid number before converting it into an integer.
> 


---

````python
# Typical Answer 6:
# Collecting Positive Numbers

numbers = []  # Creating empty list to store numbers.

while True:
    user_input = input("Enter a positive number (or type 'done' to finish): ").lower() # lower treats all inputs the same.
    
    if user_input == "done":
        break # stops and dont execute the rest code.
    elif user_input.isdigit():  # .isdigit() checks if the input contains only digits (0–9).
        number = int(user_input)
        if number > 0:
            numbers.append(number) # adding the input to the end of the list.
        else:
            print("Please enter a positive number.")
    else:
        print("Invalid input. Please enter a positive number or 'done'.")

# After input ends:
if numbers:  # if the list is not empty
    print("Numbers entered:", numbers)
    
    # Find the first even number
    first_even = None # Creates a variable to store the first even number found.
    for num in numbers:
        if num % 2 == 0:
            first_even = num
            break
            
    if first_even is not None:
        print("First even number:", first_even)
    else:
        print("No even numbers found")
else:
    print("No numbers were entered")
````


21. **Password cracker** - write a program to print the accuracy of password guess.

For example: The actual password is `MDX123` while a user tried `MDX321` or `mdx123`, the program should tell that how accurately the password was guessed, in this case it should be `50%` accuracy. if a user enter `MDX124`, the output should be `83%` and in case of `123MDX` the output should be `0%`.

Points to keep in mind while calculating the accuracy a) case of the letters entered b) order of the letters

If a password is accurately guessed, the output should state `Password cracked!`

````python
# Excellent Answer 4: 

# Taking input for the actual password
actual_password = input("Enter the actual password please: ")

while True:  # Loop until the password is cracked or the user quits
    guess_password = input("Enter your guess please (or type 'quit' to exit): ")
    
    if guess_password.lower() == "quit":  # Allow user to exit the loop
        print(" You Exitied the program. Goodbye!")
        break

    matches = 0  # Variable to count correctly matched characters
    length = len(actual_password) # Use actual password length for accuracy

    # Loop through each character position
    for i in range(length):
        # Check if the current index exists in the guess and matches the actual password
        if i < len(guess_password) and actual_password[i] == guess_password[i]:
            matches += 1  # Increment matches if characters are the same

    # Calculate accuracy (avoid division by zero)
    if length > 0:
        accuracy = round(matches / length * 100)
    else:
        accuracy = 0

    # Print result
    if accuracy == 100:
        print("Password cracked!")
        break  # Exit the loop if password is correctly guessed
    else:
        print(f"Accuracy: {accuracy}%\n")  # Show the accuracy and continue loop
````
